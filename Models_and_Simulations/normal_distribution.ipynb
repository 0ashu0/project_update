```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate data points for the standard normal distribution
x = np.linspace(-4, 4, 1000)  # Range of values for the standard normal distribution
pdf = norm.pdf(x)            # Probability density function (PDF)
cdf = norm.cdf(x)            # Cumulative distribution function (CDF)

# Plot the PDF and CDF
plt.figure(figsize=(12, 6))

# Plot the PDF
plt.subplot(1, 2, 1)
plt.plot(x, pdf, label='PDF')
plt.title('Standard Normal Distribution - PDF')
plt.xlabel('x')
plt.ylabel('Density')
plt.grid()
plt.legend()

# Plot the CDF
plt.subplot(1, 2, 2)
plt.plot(x, cdf, label='CDF', color='orange')
plt.title('Standard Normal Distribution - CDF')
plt.xlabel('x')
plt.ylabel('Probability')
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()

# Demonstrate PPF (Percent-Point Function, inverse of CDF)
quantiles = [0.025, 0.5, 0.975]  # 2.5%, 50%, and 97.5% quantiles
ppf_values = norm.ppf(quantiles)
print(f"PPF for quantiles {quantiles}: {ppf_values}")

# Verify with CDF
cdf_values = norm.cdf(ppf_values)
print(f"CDF of PPF values: {cdf_values}")
```
